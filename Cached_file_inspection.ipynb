# In this short notebook we will explore for a random year and a random race which datas we actually have 

```python
import os
import pickle

base_path = "/Users/jacoposinigaglia/Desktop/svizzera/terzo_semestre/Advanced_Data_Analysis/Progetto/cache"

print(os.listdir(base_path))
```

```python
year = "2023"
gp = "2023-03-19_Saudi_Arabian_Grand_Prix"
race = "2023-03-19_Race"

path_to_race = os.path.join(base_path, year, gp, race)
print(path_to_race)
print(os.path.exists(path_to_race))
```

```python

data_summary = {}

for f in os.listdir(path_to_race):
    if f.endswith(".ff1pkl"):
        fpath = os.path.join(path_to_race, f)
        with open(fpath, "rb") as file:
            data = pickle.load(file)
        data_summary[f] = type(data)

for k, v in data_summary.items():
    print(f"{k}: {v}")
```

```python
keys = list(dataa.keys())
print(keys)
```

```python
key = keys[18]  
driver_data = dataa[key]

print("Type:", type(driver_data))

if hasattr(driver_data, "head"):
    # It's a DataFrame
    display(driver_data)
elif isinstance(driver_data, dict):
    # Nested dict
    print("Subkeys:", list(driver_data.keys())[:10])
    pprint.pprint(driver_data[list(driver_data.keys())[0]])
else:
    # Some other type (list, array)
    pprint.pprint(driver_data)
```
