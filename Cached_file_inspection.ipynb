</center></h1>

```python
import os
import pickle

# Use relative path from project root
base_path = "cache"

print(os.listdir(base_path))
```

```python
year = "2024"
gp = "2024-09-01_Italian_Grand_Prix"
race = "2024-09-01_Race"

path_to_race = os.path.join(base_path, year, gp, race)
print(path_to_race)
print(os.path.exists(path_to_race))
```

```python

data_summary = {}

for f in os.listdir(path_to_race):
    if f.endswith(".ff1pkl"):
        fpath = os.path.join(path_to_race, f)
        with open(fpath, "rb") as file:
            data = pickle.load(file)
        data_summary[f] = type(data)

for k, v in data_summary.items():
    print(f"{k}: {v}")
```
